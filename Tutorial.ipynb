```python
import opendbpy as odb
```

```python
lef_file_path = "designs/gscl45nm.lef"
def_file_path = "designs/design.def"
```

```python
# Create the database
db = odb.dbDatabase.create()
```

```python
# Read the LEF file
odb.read_lef(db, lef_file_path)
# Read the DEF file
odb.read_def(db, def_file_path)
```

```python
chip = db.getChip()
block = chip.getBlock()
nets = block.getNets()

print("Network name | isSpecial")
print("------------------------")
for net in nets:
    print(net.getName(), net.isSpecial())
```

```python
# Iterate over all instances in the top block
for inst in block.getInsts():
    # Get the master cell of this instance
    cell = inst.getMaster()
    
    # Get cell name
    cell_name = cell.getName()
    
    # Get cell width and height
    cell_width = cell.getWidth()
    cell_height = cell.getHeight()
    
    # Calculate approximate area (this might not be exact for all cell shapes)
    cell_area = cell_width * cell_height
    
    # Print cell name and dimensions
    print(f"Cell: {cell_name}")
    print(f"Width: {cell_width}, Height: {cell_height}")
    print(f"Approximate Area: {cell_area}")
    
    # Print pin information
    print("Pins:")
    for mterm in cell.getMTerms():
        pin_name = mterm.getName()
        pin_direction = mterm.getIoType()
        print(f"  {pin_name}: {pin_direction}")
    
    print("------------------------")
```
